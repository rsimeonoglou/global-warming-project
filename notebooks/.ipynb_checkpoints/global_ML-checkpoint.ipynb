```python
import pandas as pd
import datetime as dt
from datetime import timedelta

pd.set_option("display.max_rows", 50, "display.max_columns", None)


```

```python
global_df = pd.read_csv("GlobalTemperatures.csv")
global_df
```

```python
global_df['dt'] = pd.to_datetime(global_df['dt'])
global_df['Year'] = global_df["dt"].dt.year
global_df['Month'] = global_df["dt"].dt.month
global_df['Day'] = global_df["dt"].dt.day

#datetime changed 1900-1970 to 2000s, need to correct
df_before_1900 = global_df.iloc[:1800]
df_after_1970 = global_df.iloc[2652:]

df_1900_to_1970 = global_df.iloc[1800:2652]         
df_1900_to_1970["Year"] = df_1900_to_1970["Year"]-100
df_1900_to_1970['dt'] = df_1900_to_1970["Year"].astype(str) + "-" + df_1900_to_1970["Month"].astype(str) + "-" + df_1900_to_1970["Day"].astype(str)
df_1900_to_1970['dt'] = pd.to_datetime(df_1900_to_1970['dt'])

global_df = pd.concat([df_before_1900,df_1900_to_1970,df_after_1970],ignore_index=True)
global_df
```

```python
global_df = global_df.loc[(global_df["Year"] > 1792)]
global_df = global_df.loc[(global_df["Year"] < 2014)]
global_df
```

```python
#celcius to F calc (0°C × 9/5) + 32 

global_df["land_avg_fahrenheit"] = (global_df["LandAverageTemperature"] * 9/5) +32
global_df.reset_index(inplace=True) 
global_df
```

```python
global_df_groupby = global_df.groupby("Year").mean()
global_df_groupby = global_df_groupby[["LandAverageTemperature","land_avg_fahrenheit"]].reset_index()
global_df_groupby.to_csv("global_data_for_ML.csv",index=False)
global_df_groupby = global_df_groupby.loc[(global_df_groupby["Year"]>1949)]
global_df_groupby
```

```python
X = global_df_groupby["Year"].values.reshape(-1, 1)
y = global_df_groupby["land_avg_fahrenheit"].values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
from sklearn.model_selection import train_test_split



X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score


predicted = model.predict(X_test)


mse = mean_squared_error(y_test, predicted)
r2 = r2_score(y_test, predicted)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2 ): {r2}")
```

```python
model.score(X_test, y_test)
```

```python
predicted_years = list(range(2013,2101))
predicted_temp = []
predicted_year = []

for year in predicted_years:    
    predictions = model.predict([[year]])
    prediction = predictions[0][0]
#     print(prediction)
    predicted_temp.append(prediction)
    predicted_year.append(year)

predicted_temp
predicted_year
```

```python
predicted_df = pd.DataFrame({"Year":predicted_year,"land_avg_fahrenheit":predicted_temp})
predicted_df.to_csv("predicted_global.csv")
```

```python
import pandas as pd
import io
import requests
url="https://raw.githubusercontent.com/owid/co2-data/master/owid-co2-data.csv"
s=requests.get(url).content
c=pd.read_csv(io.StringIO(s.decode('utf-8')))
```

```python
groupby = c.groupby("year").sum().reset_index()
groupby_filter = groupby.loc[(groupby["year"]>1949) &
                            (groupby["year"]<2013)]

groupby_filter=groupby_filter.rename(columns = {"year":"Year"})
groupby_filter
```

```python
X = groupby_filter["Year"].values.reshape(-1, 1)
y = groupby_filter["co2"].values.reshape(-1, 1)

print("Shape: ", X.shape, y.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X, y)
```

```python
from sklearn.model_selection import train_test_split



X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
predicted_years = list(range(2013,2101))
predicted_co2 = []
predicted_year = []

for year in predicted_years:    
    predictions = model.predict([[year]])
    prediction = predictions[0][0]
#     print(prediction)
    predicted_co2.append(prediction)
    predicted_year.append(year)

predicted_co2
predicted_year
```

```python
predicted_co2 = pd.DataFrame({"Year":predicted_year,"co2":predicted_co2})
predicted_co2.to_csv("predicted_co2.csv")
predicted_co2
```

```python
#check to actual
actual_values = groupby.loc[groupby["year"]>2012]
actual_values = actual_values.rename(columns = {"year":"Year"})
check_values = pd.merge(predicted_co2,actual_values,on="Year")
check_values = check_values[['Year','co2_x','co2_y']]
check_values = check_values.rename(columns = {"co2_x":"predicted_co2","co2_y":"actual_co2"})

check_values["Difference"] = check_values["predicted_co2"]- check_values["actual_co2"] 
check_values
```
