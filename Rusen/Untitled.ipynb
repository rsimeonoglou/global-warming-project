```python
import pandas as pd
```

```python
population_predict = pd.read_csv("Population_predicted_ML.csv")
population_predict = population_predict.loc[population_predict["Year"]>2010]
population_predict
```

```python
beg_pop = 0
population_predict["year change"] = 0
population_predict["running_total_change"] = 0

for index, row in population_predict.iterrows():
    if row["Year"] == 2015:
        population_predict.loc[index, 'year change'] = 0
        population_predict.loc[index, 'running_total_change'] = 0
        beg_pop = row["population_predicted"]
        running_total = 0
    else:
        change = row["population_predicted"] - beg_pop
        running_total = running_total + change
        population_predict.loc[index, 'year change'] = change
        population_predict.loc[index, 'running_total_change'] = running_total
        beg_pop = row["population_predicted"]
        
population_predict.head(20)
```

```python
population_predict.to_csv("predicted_population_year_change.csv",index=False)
```
